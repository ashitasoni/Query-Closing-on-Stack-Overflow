```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
offTopic = pd.read_csv('102_OffTopic.csv')
unclear = pd.read_csv('103_Unclear.csv')
tooBroad = pd.read_csv('104_TooBroad.csv')
opinionBased = pd.read_csv('105_OpinionBased.csv')
openQues = pd.read_csv('open.csv')
```

```python
closeQues = pd.concat([offTopic, unclear, tooBroad, opinionBased])
```

```python
closeQues.columns
```

```python
openQues.columns
```

```python
openQues['experienceInTime'] = -1*openQues['experienceInTime']
# print(min(openQues['experienceInTime']))
```

```python
maxExpC = max(closeQues['experienceInTime'])
minExpC = min(closeQues['experienceInTime'])
avgExpC = np.mean(closeQues['experienceInTime'])
print('For Closed Questions')
print('Maximum Experience: ' + str(maxExpC))
print('Minimum Experience: ' + str(minExpC))
print('Mean Experience: ' + str(avgExpC))
```

```python
maxExpO = max(openQues['experienceInTime'])
minExpO = min(openQues['experienceInTime'])
avgExpO = np.mean(openQues['experienceInTime'])
print('For Open Questions')
print('Maximum Experience: ' + str(maxExpO))
print('Minimum Experience: ' + str(minExpO))
print('Mean Experience: ' + str(avgExpO))
```

```python
plt.figure(figsize=(20,15))
plt.grid(True)
plt.title('User\'s Experience vs No. of Closed and Open Questions')
plt.hist([closeQues['experienceInTime'], openQues['experienceInTime']], bins = 50, rwidth = 0.5, orientation = u'horizontal', label=['closed', 'open'])
plt.xlabel('# Questions')
plt.ylabel('User Experience in time')
plt.legend()
plt.show()
```

```python
maxUpvotesC = max(closeQues['Upvotes'])
minUpvotesC = min(closeQues['Upvotes'])
avgUpvotesC = np.mean(closeQues['Upvotes'])
print('For Closed Questions')
print('Maximum Upvotes: ' + str(maxUpvotesC))
print('Minimum Upvotes: ' + str(minUpvotesC))
print('Mean Upvotes: ' + str(avgUpvotesC))
```

```python
maxUpvotesO = max(openQues['Upvotes'])
minUpvotesO = min(openQues['Upvotes'])
avgUpvotesO = np.mean(openQues['Upvotes'])
print('For Open Questions')
print('Maximum Upvotes: ' + str(maxUpvotesO))
print('Minimum Upvotes: ' + str(minUpvotesO))
print('Mean Upvotes: ' + str(avgUpvotesO))
```
