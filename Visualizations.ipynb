```python
import matplotlib.pyplot as plt
import pandas as pd
import re
```

```python
openCloseData=pd.read_csv("openCloseData.csv")
```

```python
openCloseData.shape
```

```python
plt.hist( openCloseData[openCloseData['closed']==1]['tagCount'] ,label="Closed") 
plt.hist( openCloseData[openCloseData['closed']==0]['tagCount'] ,label="Open") 
plt.xlabel("Count of Tags in a post")
plt.ylabel("Frequency")
plt.title("Histogram of frequency of count of tags in open and closed questions")
plt.legend()
plt.show()
# It can be seen that more proportion of closed questions(as compared to open questions) have lower count of tags(in range [1,2])
# while more proportion of open questions(as compared to closed questions) have higher count of tags(in range [3,5])
```
